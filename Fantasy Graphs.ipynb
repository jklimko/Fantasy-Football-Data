```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
data_filepath = "/Users/jtklimko/Desktop/Data/data_v2/yearly/2019.csv"
fantasy_data = pd.read_csv(data_filepath)
```

```python
#fantasy_data.head()
top_scorers = fantasy_data.head(30)
top_scorers.sort_values(by=['FantasyPoints'], inplace=True, ascending=False)
```

```python
top_rbs = fantasy_data[fantasy_data.Pos == 'RB']
top_rbs = top_rbs.head(30)
top_rbs.sort_values(by=['FantasyPoints'], inplace=True, ascending=False)
```

```python
top_wrs = fantasy_data[fantasy_data.Pos == 'WR']
top_wrs = top_wrs.head(30)
top_wrs.sort_values(by=['FantasyPoints'], inplace=True, ascending=False)
```

```python
top_tes = fantasy_data[fantasy_data.Pos == 'TE']
top_tes = top_tes.head(30)
top_tes.sort_values(by=['FantasyPoints'], inplace=True, ascending=False)
```

```python
plt.figure(figsize = (16,6))
plt.xticks(ha="left",rotation = -45)
plt.grid(b=True)
sns.barplot(x=top_rbs['Player'], y=top_rbs['FantasyPoints'])
```

```python
plt.figure(figsize = (16,6))
plt.xticks(ha="left",rotation = -45)
plt.grid(b=True)
sns.barplot(x=top_wrs['Player'], y=top_wrs['FantasyPoints'])
```

```python
plt.figure(figsize = (16,6))
plt.xticks(ha="left",rotation = -45)
plt.grid(b=True)
sns.barplot(x=top_tes['Player'], y=top_tes['FantasyPoints'])
```
